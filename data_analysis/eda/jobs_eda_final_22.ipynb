```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# CSV 파일 읽어오기
csv_file_22 = 'E:/project/devcourse_DW_BI_project/data/2022_jobs.csv'

df_22 = pd.read_csv(csv_file_22, encoding='utf-8')
header_22 = df_22.columns.tolist()
```

```python
duplicates = df_22.duplicated()
print(duplicates)
```

```python
missing_values_22= df_22.isnull().sum()
print("[결측치 확인]")
print(missing_values_22)
for i, v in missing_values_22.items():
    print(v)
```

```python
total_count = len(df_22['workPlc'])
print("null 포함 전체 개수 : ", total_count )

null_count = df_22['workPlc'].isnull().sum()
print("null 개수 : ", null_count )
```

```python
t = df_22['workPlc'].value_counts()
print("길이 : ", len(t))
for i, v in t.items():
    print(i, v)
```

## 근무지명 

#### 결측치

```python
missing_values_22['workPlcNm']
```

```python
total_count = len(df_22['workPlcNm'])
print("null 포함 전체 개수 : ", total_count )

null_count = df_22['workPlcNm'].isnull().sum()
print("null 개수 : ", null_count )
```

```python
t_workPlc = df_22['workPlcNm'].value_counts()
도단위 = {}
print("길이 : ", len(t_workPlc))
for i, v in t_workPlc.items():
    print(i, v)
#     print(str(i).split()[0])
    도 = str(i).split()[0]
    도단위[도] = 도단위.get(도, 0) + v
```

```python
print(도단위)
print(sum(도단위.values()))
```
