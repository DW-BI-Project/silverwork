####  "projTypeCd"별로 "projTypeNm"의 종류 확인

### 칼럼별 null 및 카테고리 확인

```python
# 칼럼별 null 및 카테고리 확인
for h in header : 
    print("\n\n|------------------", h)
    total_count = len(df[h])
    print("null 포함 전체 개수 : ", total_count )

    null_count = df[h].isnull().sum()
    print("null 개수 : ", null_count )

    count = df[h].notnull().sum()
    print("null 제외 전체 개수 : ", count)

    value_counts = df[h].value_counts()
    print(value_counts)
    print(len(value_counts))
```
